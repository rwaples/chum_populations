```python
chum_map = pd.read_csv('/home/ipseg/Desktop/waples/chum_populations/linkage_map/LEPmap/with_paralogs/final/PS_chum_map_2015.txt', 
                       header = 0, sep = '\t')
chum_map.head()
```

```python
locus_order = pd.read_csv('/home/ipseg/Desktop/waples/chum_populations/data/batch_4/Ne/mapped_non_paralogs.map', 
                       header = None, sep = '\t')
locus_order.columns = ['other_LG', 'stacks_SNP', 'junk1', 'junk2']
locus_order = locus_order[['other_LG', 'stacks_SNP']]
locus_order['LDNe_ID'] = [xx+1 for xx in locus_order.index.tolist()]
locus_order.head()
```

```python
merge1 = pd.merge(locus_order, chum_map)
merge1.sort(columns = ['paper1_LG', 'cM'], inplace=True)
merge1.head()
```

```python
order_of_LDNe_IDs = [x for x in merge1['LDNe_ID']]
pos_of_LDNe_IDs = dict(zip(range(1, len(order_of_LDNe_IDs)+1), order_of_LDNe_IDs))
```

```python
r2_df['Loc1_pos'] = [pos_of_LDNe_IDs[xx] for xx in r2_df['Loc1'] ]
r2_df['Loc2_pos'] = [pos_of_LDNe_IDs[xx] for xx in r2_df['Loc2'] ]

r2_df.head()
```

```python
r2_df_mat = r2_df.pivot(index = 'Loc1_pos', columns = 'Loc2_pos', values = 'Mean_r^2')
```

```python
r2_df_mat.head()
```

```python
sns.jointplot(r2_df_mat[5200], r2_df_mat[1300], kind = 'hex')
```
