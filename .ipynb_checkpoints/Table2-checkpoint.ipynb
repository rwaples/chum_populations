## Ne

```python
Ne = pd.read_csv(os.path.join('results','batch_4','Ne', 'Ne.txt'), sep = '\t')
Ne
```

```python
## Replace negative and infinite values with 10,000
```

```python
Ne = Ne.replace('Infinite', np.NaN)
num = Ne._get_numeric_data()
num[num < 0] = 10000   
Ne
```

## Heterozygosity

```python
hetmiss = pd.read_csv(os.path.join('paper','tables','hetmiss.txt'), sep = '\t')
hetmiss.head()
```

```python
table_2_raw = pd.merge(Ne, hetmiss)
table_2_raw.head()
```

```python
table_2_raw = table_2_raw.rename(columns={'O_het':'Heterozygosity'})
```

```python
table2_grouped = table_2_raw.groupby('POPNAME')
table2_grouped.aggregate(np.average)[['Heterozygosity', 'Ne']]
```

```python
table2 = table_2_raw.groupby('POPNAME').agg([np.mean])[['Heterozygosity', 'Ne']]
#table2.loc[:, ('Aligned sequences', 'std')] = [np.int(xx) for xx in table1['Aligned sequences']['std']]
#table2.insert(0, 'n', table2_grouped.count()['Ne'])
table2
```

```python
table2.columns = table2.columns.droplevel(1)
```

```python
table2
```

```python
formatters = {'Heterozygosity':  lambda x: '%10.2f' % x,
              'Ne':  lambda x: "{:,}".format(np.int(x)),
             }
print table2.to_latex(formatters=formatters, index_names = False)
```
