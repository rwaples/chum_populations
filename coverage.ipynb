## Coverage of reference contigs

```python
import os.path
import glob
import numpy as np
import pandas as pd
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=2)
sns.set_style("white")
%matplotlib inline
```

```python
coverage_dir = '/home/ipseg/Desktop/waples/chum_populations/data/batch_4/coverage'
```

```python
cd {coverage_dir}
```

```python
bam_covs = glob.glob(os.path.join(coverage_dir , '*.bam.cov'))
```

### Load coverage files

```python
def reader(f):
    d = pd.read_csv(f, header=None, sep = "\t")
    d.columns = ['contig', 'start', 'stop', 'depth']
    ind_ID = os.path.basename(f).split(".")[0]
    d['ID'] = ind_ID
    d['SILLI'] = ind_ID.split("_")[0]
    return(d)
```

```python
cov = pd.concat([reader(f) for f in bam_covs])
cov.head()
#len(cov)/174
```

### Mean coverage of contigs 
This is the input data to Stacks (pstacks)

```python
figsize(8,8)
sns.distplot(np.log10(cov.groupby('contig').mean()['depth']))
plt.xlabel("log10(mean depth)")
sns.despine()
plt.show()
```

# Depth of genotypes

### Reformat genotypes in the *.raw format into a long format - 
'melt' them - one line per genotype

```python
calls = pd.read_csv(
    '/home/ipseg/Desktop/waples/chum_populations/data/batch_4/pop_genotypes/non_paralogs.raw',
    sep = " ")
calls.columns = list(calls.columns[0:6])+ [xx[:-2] for xx in calls.columns[6:]]
calls_melt = pd.melt(calls, id_vars = list(calls.columns.values[0:6]), 
        var_name = 'stacks_SNP', value_name = 'call')
calls_melt.head()
```

```python
SNP_contigs = pd.read_csv('/home/ipseg/Desktop/waples/chum_populations/data/batch_4/batch_4.plink.map', 
                        sep ='\t', header = None, comment = "#")
SNP_contigs.columns = ['contig', 'stacks_SNP', 'pos1', 'pos2']
SNP_contigs.head()
```

```python
calls_join = pd.merge(calls_melt, SNP_contigs, on = 'stacks_SNP', how = 'left')
calls_join.head()
#len(calls_join)/174
```

#### combine with coverage data
I replace depth above 200 with 201 so I can do away with the log scaling.

```python
calls_depth = pd.merge(calls_join, cov, how = 'left',
                       left_on = ['IID', 'contig'], right_on = ['ID', 'contig'])

calls_depth['call'].fillna(value = -9, inplace = True)
calls_depth['depth'].fillna(value = 0, inplace = True)
calls_depth['depth'] = calls_depth['depth'].where(calls_depth['depth'] < 200, 201)
#len(calls_depth)/174
```

Depth distribution for genotypes/loci passing filters 

```python
sns.distplot(calls_depth['depth'], bins = 201)
sns.despine()
plt.show()
```

#### Depth by genotype call:
    0 = homozygous
    1 = heterozygous
    2 = homozygous
    -9 = missing
Inner lines show median and 75% & 25% quantiles. Notice higher depth on average for het calls.

```python
figsize(6,6)
sns.violinplot(calls_depth['depth'], groupby = calls_depth['call'])
sns.despine()
plt.ylim(0, 100)
plt.show()
```

### Depth per population

```python
figsize(8, 8)
sns.violinplot(calls_depth['depth'], groupby = calls_depth['FID'])
plt.ylim(0, 120)
sns.despine()
plt.show()
```

### Individual mean depth vs \# of called genotypes

```python
from __future__ import division
ind_depth = calls_depth.groupby('IID').mean()['depth']
ind_nocalls = calls_depth[calls_depth['call'] >-1].groupby('IID').count()['call']
ind_het = calls_depth[calls_depth['call'] == 1].groupby('IID').count()['call']/(
    calls_depth[calls_depth['call'] >-1].groupby('IID').count()['call'])
```

```python
sns.jointplot(ind_depth, ind_nocalls, size = 8, kind = 'scatter', stat_func = None)
plt.show()
```

### Heterozygosity vs depth
A lowess  (locally weighted scatterplot smoothing) line is fit to this relationship

```python
sns.regplot(ind_depth, ind_het, lowess=True)
plt.ylabel('heterozygosity')
plt.xlim(0,100)
sns.despine()
plt.show()
```
