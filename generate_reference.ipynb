### remove self alignments
sam flag 256 = secondary alignment

```python
SAM_filename = "/media/Shared/Data/chum/populations/ref/paper_1/chum_ref.self.sam"
SAM_self_removed_filename = SAM_filename.replace('self', 'self_removed')

# takes a basic SAM file and removes lines representing the self alignment (readName == refName)
with open(SAM_filename) as INFILE:
    with open(SAM_self_removed_filename, 'w') as OUTFILE:
        for line in INFILE:
            if line.startswith("@"):
                pass
            else:
                (readName, sumFlags, refName, refPos, mapQ, CIGAR, mate, matePos, fragmentLen, readSeq, readQ, alignScore), otherScores  = line.strip().split("\t")[:12], line.strip().split("\t")[12:] 
                if readName != refName:
                    OUTFILE.write(line)
```

### blacklist all but one of a group of catalog entries that align

```python
blacklist = set()
with open(SAM_self_removed_filename) as INFILE:
    for line in INFILE:
        if line.startswith("@"):
            pass
        else:
            readName, sumFlags, refName, refPos, mapQ, CIGAR, mate, matePos, fragmentLen, readSeq, readQ, alignScore, XS, XN, XM, XO, XG, NM, MD, YT = line.strip().split("\t")
            if readName in whitelist and refName in whitelist:
                pass
            elif refName in whitelist and np.int(NM.split(":")[2]) <= 5:
                blacklist.add(readName)
            elif readName in whitelist and np.int(NM.split(":")[2]) <= 5:
                blacklist.add(refName)
            elif readName not in whitelist and refName not in whitelist and readName not in blacklist:
                whitelist.add(readName)
                blacklist.add(refName)
```

### Look at our white and black lists

```python
print('length of whitelist: {}'.format(len(whitelist)))
print('length of blacklist: {}'.format(len(blacklist)))
print('intersecting loci: {}'.format(len(whitelist.intersection(blacklist))))
```

```python
whitelist_file = SAM_filename + ".whitelist"
blacklist_file = SAM_filename + ".blacklist"

with open(blacklist_file, 'w') as BLACKLIST:
   BLACKLIST.writelines([(str(x)+"\n") for x in blacklist])
    
with open(whitelist_file, 'w') as WHITELIST:
   WHITELIST.writelines([(str(x)+"\n") for x in whitelist])
```
