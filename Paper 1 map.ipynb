```python
import pysam
import os.path
import numpy as np
import pandas as pd
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=2)
sns.set_style("white")
%matplotlib inline
```

```python
from Bio import SeqIO
```

```python
chum_map = pd.read_csv(os.path.join('linkage_map', 'LEPmap', 'with_paralogs', 'final', 'PS_chum_map_2015.txt'), 
                       sep = '\t', header = 0)
chum_map.head()
```

```python
chum_ref = '/media/Shared/Data/chum/populations/ref/paper_1/chum_paper_1_ref.CURATED.fasta'
```

```python
fasta_sequences = SeqIO.parse(open(chum_ref),'fasta')
```

```python
seq_of_ID = dict([(xx.name, [str(xx.seq)]) for xx in fasta_sequences])
seq_of_ID
```

```python
seqs = pd.DataFrame.from_dict(seq_of_ID, orient = 'index')
seqs.columns = ['seq']
seqs.head()
```

```python
map1 = pd.read_csv('/home/ipseg/Desktop/waples/temp/Map1.txt', sep = '\t')
map1.head()
```

```python
map1['contig'] = ['c'+xx.split('_')[0] for xx in map1['locus'] ]
map1.head()
```

```python
merged = pd.merge(map1,seqs, left_on= 'contig', right_index=True)
merged = merged[['contig', 'locus', 'LG', 'cm', 'seq']]
merged.head()
```

```python
merged.to_csv('/home/ipseg/Desktop/waples/temp/Map1-revised.txt', index = False, sep = '\t')
```
