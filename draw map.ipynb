```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import os.path
import numpy as np
import pandas as pd
import seaborn as sns
from IPython.display import Image
import matplotlib.cm as cm
from IPython.core.pylabtools import figsize
sns.set(font_scale=2)
sns.set_style("white")
%matplotlib inline
```

```python
POPINFO = pd.read_csv(os.path.join('data', 'POPINFO.latlong.txt.csv'), sep = '\t')
POPINFO.head()
```

```python
max_lat = max(POPINFO['LAT'])
min_lat = min(POPINFO['LAT'])
max_long = max(POPINFO['LONG'])
min_long = min(POPINFO['LONG'])
(min_lat, max_lat), (min_long, max_long)
```

```python
my_map = Basemap(projection='merc', lat_ts = 48,
    resolution = 'h', area_thresh = 0.9,
    llcrnrlon=-125, llcrnrlat=47.0,
    urcrnrlon=-121.5, urcrnrlat=49.5)
```

```python
figsize(10,10)
```

```python
color_by = 'TIMING'

my_map.drawcoastlines()
#my_map.drawcountries()
my_map.fillcontinents(color = 'gray', lake_color = 'white', alpha = .2)
my_map.drawmapboundary(fill_color='white')

#mycolors = cm.rainbow(np.linspace(.5, 1, len(set(POPINFO[color_by]))))
mycolors = sns.color_palette("cubehelix", len(set(POPINFO[color_by])))

mymarkers = ['o', '^', 's']

for idx, xx in enumerate(set(POPINFO[color_by])):
    subset = POPINFO[POPINFO[color_by] == xx]
    x, y = my_map(list(subset['LONG']), list(subset['LAT']))
    plt.plot(x, y, mymarkers[idx], 
             #linewidth = 20,
             #color = mycolors[idx], 
             #edgecolors = 'purple',
             color = 'black', 
             markersize=20, label = xx, alpha = .8)
legend = plt.legend(loc="lower left", bbox_to_anchor=(0,0), title = 'Run timing')
legend.get_title().set_fontsize('24')


plt.savefig(os.path.join('paper', 'figures', 'collection_map.png'))
plt.show()
```
