```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import os.path
import numpy as np
import pandas as pd
import seaborn as sns
from IPython.display import Image
import matplotlib.cm as cm
from IPython.core.pylabtools import figsize
sns.set(font_scale=2)
sns.set_style("white")
%matplotlib inline
```

```python
from  mpl_toolkits.axes_grid1 import inset_locator
```

```python
pop_colors = ['purply pink', 'golden', 'leaf green', 'blood', 'vomit', 'primary blue', 
              'bright sky blue', 'dark cyan', 'grey', 'red']
```

```python
sns.palplot(sns.xkcd_palette(pop_colors))
```

```python
[pop_colors[xx-1] for xx in [1,7,2,4,5,6,9,10,8,3]]
```

```python
POPINFO = pd.read_csv(os.path.join('data', 'POPINFO.latlong.txt.csv'), sep = '\t')
POPINFO = POPINFO.sort_values('CLUSTER')
POPINFO['map_order'] = [1,7,2,4,5,6,9,10,8,3]
POPINFO = POPINFO.sort_values('map_order')
POPINFO
```

```python
max_lat = max(POPINFO['LAT'])
min_lat = min(POPINFO['LAT'])
max_long = max(POPINFO['LONG'])
min_long = min(POPINFO['LONG'])
(min_lat, max_lat), (min_long, max_long)
```

```python
my_map = Basemap(projection='merc', lat_ts = 48,
    resolution = 'h', area_thresh = 0.9,
    llcrnrlon=-125, llcrnrlat=47.0,
    urcrnrlon=-121.5, urcrnrlat=49.5)
```

```python
POPINFO.sort_values('map_order')[['POPNAME', 'TIMING']]
```

```python
figsize(10,10)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)


color_by = 'TIMING'

my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color = 'gray', lake_color = 'white', alpha = .2)
my_map.drawmapboundary(fill_color='white')

#mycolors = cm.rainbow(np.linspace(.5, 1, len(set(POPINFO[color_by]))))
mycolors = sns.xkcd_palette(pop_colors)

mymarkers = ['o', 's', 's', 's', 'o', 's', '^', 's', 's', 's']

for idx, xx in enumerate(POPINFO.sort_values('map_order')['POPNAME']):
    subset = POPINFO[POPINFO['POPNAME'] == xx]
    x, y = my_map(list(subset['LONG']), list(subset['LAT']))
    plt.plot(x, y, mymarkers[idx], 
             #linewidth = 20,
             #color = mycolors[idx], 
             markeredgecolor = 'black', markeredgewidth = 2,
             #color = 'black', 
             color = mycolors[idx],
             markersize=17, label = xx, alpha = .8)
legend = plt.legend(bbox_to_anchor=(1.60,.8), title = 'Collection')
legend.get_title().set_fontsize('24')


x_usa, y_usa = my_map(-122, 48.920)
x_can, y_can = my_map(-122, 49.020)
plt.text(x_usa, y_usa, 'USA', size = 20)
plt.text(x_can, y_can, 'Canada', size = 20)

#inset
axins = inset_locator.inset_axes(parent_axes = ax, width = 2, height = 2, loc = 2, 
                                        bbox_to_anchor=(0.138, .941), 
                                        bbox_transform=plt.gca().figure.transFigure
                                       )

insetmap = Basemap(llcrnrlon=-135,
                   llcrnrlat=20,
                   urcrnrlon=-80,
                   urcrnrlat=60, 
                   ax=axins)

insetmap.drawcoastlines()
insetmap.drawcountries()
insetmap.fillcontinents(color = 'gray', lake_color = 'white', alpha = .2)
insetmap.drawmapboundary(fill_color='white')

plt.plot(-122, 48.920,'s', 
             markeredgecolor = 'black', markeredgewidth = 2,
             color = 'gray', markerfacecolor='None',
             markersize=25, label = xx, alpha = 1)


plt.savefig(os.path.join('paper', 'figures', 'collection_map_colors.png'), bbox_inches='tight', dpi = 300)
plt.show()
```
